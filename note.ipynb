**Task 4.Data Cleaning**

```python
data.isnull().sum()

#droping all missing Values
data.dropna(inplace=True)
#filling missing values
#data.fillna(0,inplace=True)
```

```python
# find the duplicate
data.duplicated().sum()

#to see the the duplicated
#data[data.duplicated()]

data.drop_duplicates(inplace=True)
data.duplicated().sum()
data[data.duplicated()]
```

### EDA
**Task 5. Data Analysis**

**Univariate Analysis**

```python
#identify the outliers in price
df = data[data['price'] < 1500]
sns.boxplot(data=df,x='price')
plt.show()
```

```python
#histogram
plt.figure(figsize=(8,5))
sns.histplot(data=df,x='price',bins=100)
plt.title("Price Distribution")
plt.ylabel("Frequency")
plt.show()
```

```python
data.dtypes
```

```python
plt.figure(figsize=(6,3))
sns.histplot(data=df,x='availability_365')
plt.title("availability_365 Price Distribution")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6, 3))
sns.countplot(data=df, x='price', hue='beds')
plt.title("Availability by Number of Beds")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6, 3))
sns.histplot(data=df, x='room_type', y='reviews_per_month', bins=10)
plt.title("room_type vs. Reviews per Month")
plt.ylabel("Reviews per Month")
plt.show()

```

```python
plt.figure(figsize=(6, 3))
sns.scatterplot(data=df, x='room_type', y='reviews_per_month')
plt.title("room_type vs. Reviews per Month")
plt.ylabel("Reviews per Month")
plt.show()
```

```python
df.groupby(by='neighbourhood_group')['price per beds'].mean()
```

```python
df.groupby(by='neighbourhood_group')['price'].mean()
```

**Feature Engineering**

```python
# Creating new column
#price per bed

df['price per beds']=df['price']/df['beds']
df.head()
```

**Bi variable Analysis**
one variable dependency on another variables


```python
df.columns
```

```python
#Relation between neighbourhood_group with price
sns.barplot(data=df,x='neighbourhood_group', y='price', hue='room_type')
plt.show()
```

```python
#No of reviews with price relationships
sns.scatterplot(data=df,x='number_of_reviews',y='price',hue='neighbourhood_group')
plt.show()
```

```python
df.dtypes
```

```python

sns.pairplot(data=df,vars=['price','availability_365','minimum_nights','number_of_reviews'],hue='room_type')
plt.show()
```

```python
#Geographical Distribution of AirBnb Listing
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df,x='latitude',y='longitude',hue='room_type')
plt.title('Geographical Distribution of AirBnb Listing')
plt.show()
```

```python
# Filter numeric columns
numeric_columns = df.select_dtypes(include=['float64', 'int64'])

# Generate the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_columns.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```
